```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
import seaborn as sns
```

```python
flipkart_data = pd.read_csv('flipkart_data.csv')
flipkart_data.head()
```

```python
# Displaying all the columns of the dataframe

flipkart_data.columns
```

```python
flipkart_data.info()
```

```python
# Statistical summary of the dataframe

flipkart_data.describe()
```

```python
# Removing duplicates from the dataframe.

flipkart_data = flipkart_data.drop_duplicates() 
```

```python
# Looking for rows that have more than 2 missing values

flipkart_data[flipkart_data.isnull().sum(axis = 1) > 2]
```

```python
#  Dropping rows that have more than 2 missing values

flipkart_data = flipkart_data.dropna(thresh = 6)
```

```python
# since the Model column has only two None values so it is safe that two rows with None values.

flipkart_data = flipkart_data.dropna(subset = ['Model']) 
```

```python
# Finding the most frequent color to fill in place of null values. 

flipkart_data['Color'].mode()
```

```python
# Filling the missing values in Color column with the most frequent color.

flipkart_data['Color'] = flipkart_data['Color'].fillna('Black')
```

```python
# Dropping rows with null values in memory column

flipkart_data = flipkart_data.dropna(subset = ['Memory']) 
```

```python
# Dropping rows with null values in memory column

flipkart_data = flipkart_data.dropna(subset = ['Storage']) 
```

```python
# Filling the missing values in Selling Price column with the mean of the column.

flipkart_data['Selling Price'] = flipkart_data['Selling Price'].fillna(flipkart_data['Selling Price'].mean()) 
```

```python
# Creating numeric columns for memory and storage.

flipkart_data['Memory_gb'] = flipkart_data['Memory'].apply(
    lambda x: float(x.replace('GB','').replace('MB',''))/1024 if 'MB' in x else float(x.replace('GB',''))
)

flipkart_data['Storage_gb'] = flipkart_data['Storage'].apply(
    lambda x: float(x.split()[0])/1024 if isinstance(x, str) and 'MB' in x
    else float(''.join(filter(str.isdigit, x))) if isinstance(x, str) and any(c.isdigit() for c in x)
    else np.nan
)
```

```python
flipkart_data.head()
```

```python
# Dropping original_price column as this is having 63% of missing values.

flipkart_data = flipkart_data.drop(columns = ['Original Price'],axis = 1)
```

```python
# Adding a new column 'Discount' which is 15% of the selling price.

flipkart_data['Discount'] = flipkart_data['Selling Price'] * 0.15 
```

```python
# Final Price column after discount.

flipkart_data['Final_Price'] = flipkart_data['Selling Price'] - flipkart_data['Discount'] 
```

```python
# Displaying the info of the cleaned dataframe.

flipkart_data.info() 
```

```python
sns.countplot(data = flipkart_data, x = 'Brand',hue = 'Brand',
              order = flipkart_data['Brand'].value_counts().index)

plt.xticks(rotation = 45)
plt.title('Number of phones sold in each brand')
plt.xlabel('Brand')
plt.ylabel('Nuber of phones sold')
plt.show()
```

```python
top_10 = flipkart_data.groupby('Brand')['Selling Price'].mean()
top_10 = top_10.sort_values(ascending=False)
top_10 = top_10.head(10)

plt.barh(top_10.index,top_10.values,color = 'purple',label = 'Top 10 Brands')
plt.title('Top 10 Brands with Highest Average Price on Flipkart')
plt.xlabel('Selling Price')
plt.ylabel('Brand')
plt.legend()



plt.show()
```

```python
# Grouped Bar Chart

fig,ax = plt.subplots(1,3,figsize = (25,8))
ax = ax.flatten()

avg_data = flipkart_data.groupby('Brand')[['Rating','Final_Price','Discount']].mean()


ax[0].bar(avg_data.index,avg_data['Rating'],color = 'skyblue',label = 'Average Rating')
ax[0].set_title('Average rating for each brand')
ax[0].set_xlabel('Brand')
ax[0].set_ylabel('Rating')
ax[0].legend()
ax[0].set_xticks(range(len(avg_data.index)))
ax[0].set_xticklabels(avg_data.index, rotation=45)



ax[1].bar(avg_data.index,avg_data['Final_Price'],color = 'red',label = 'Average Price')
ax[1].set_title('Average price for each brand')
ax[1].set_xlabel('Brand')
ax[1].set_ylabel('price')
ax[1].legend()
ax[1].set_xticks(range(len(avg_data.index)))
ax[1].set_xticklabels(avg_data.index, rotation=65)

ax[2].bar(avg_data.index,avg_data['Discount'],color = 'orange',label = 'Average Discount')
ax[2].set_title('Average Discount for each brand')
ax[2].set_xlabel('Brand')
ax[2].set_ylabel('Discount')
ax[2].legend() 
ax[2].set_xticks(range(len(avg_data.index)))
ax[2].set_xticklabels(avg_data.index, rotation=45)

fig.suptitle('Grouped Bar chart for mobile brands')




plt.show() 
```

```python
# Bivariate analysis

# Detecting outliers in final_price for each brand

plt.figure(figsize = (14,5))

sns.boxplot(x = 'Brand',y = 'Final_Price',hue = 'Brand',data = flipkart_data,showmeans = True,
            meanprops = {'marker':'o','markerfacecolor':'white','markeredgecolor':'black','markersize':5},
            palette = 'Spectral') 
plt.xlabel('Brand')
plt.ylabel('Total Price')
plt.title('Price Distribution of Mobile Brands on Flipkart with Outliers')

plt.show()
```

```python
# Storage_gb Vs Memory_gb

plt.figure(figsize = (6,4))

sns.regplot(x = 'Storage_gb', y = 'Memory_gb',data = flipkart_data,
           marker = 'o',
           ci = None,
           scatter_kws = dict(color = 'skyblue',s = 4),
           line_kws = dict(linestyle = '--',color = 'red'))

plt.title('Storage_gb Vs Memory_gb')
plt.xlabel('Storage')
plt.ylabel('Memory')
plt.grid()



plt.show()
```

```python
# Multi variate analysis.

# Selecting all numeric columns for cluster map
numeric_cols = ['Memory_gb', 'Storage_gb', 'Selling Price', 'Final_Price', 'Discount', 'Rating']

# Replace any infinite values with NaN, then fill NaN with mean
flipkart_data[numeric_cols] = flipkart_data[numeric_cols].replace([np.inf, -np.inf], np.nan)
flipkart_data[numeric_cols] = flipkart_data[numeric_cols].fillna(flipkart_data[numeric_cols].mean())

# Compute correlation matrix
corr_matrix = flipkart_data[numeric_cols].corr()

# Plotting clustermap
sns.clustermap(corr_matrix, annot=True, cmap='coolwarm', figsize=(10,8),
              linewidth = 1,linecolor = 'black',
              annot_kws = dict(size = 10))
plt.title('Clustermap of Numeric Variables', pad=80)
plt.show()
```

```python
g = sns.pairplot(vars = ['Selling Price','Discount'],data = flipkart_data,hue = 'Brand',palette = 'coolwarm',
            height = 3)
g.figure.suptitle('Relationship between Selling Price & Discount')
g.figure.subplots_adjust(top = 0.9) 

plt.show() 
```
